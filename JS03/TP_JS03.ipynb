```python
df = pd.read_csv('wbc.csv')

print("üìå Info dataset:")
df.info()

print("\nüìä 5 baris pertama:")
display(df.head())

summary = pd.DataFrame({
    'dtype': df.dtypes,
    'missing_values': df.isnull().sum(),
    'unique_values': df.nunique()
})

print("\nüìã Ringkasan kolom:")
display(summary)

usable = summary[
    (summary['missing_values'] < len(df)) &
    (summary['unique_values'] > 1) &
    (~summary.index.str.contains('id', case=False))
]

not_usable = summary.loc[~summary.index.isin(usable.index)]

print("\n‚úÖ Variabel yang dapat digunakan:")
display(usable)

print("\nüö´ Variabel yang TIDAK dapat digunakan:")
display(not_usable)
```

```python
print("üîç Nilai unik sebelum encoding:")
print(df['diagnosis'].unique())

df['diagnosis_encoded'] = df['diagnosis'].map({'M': 1, 'B': 0})

print("\n‚úÖ Nilai unik setelah encoding:")
print(df['diagnosis_encoded'].unique())

display(df[['diagnosis', 'diagnosis_encoded']].head())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()

numerical_cols = [col for col in numerical_cols if col != 'diagnosis_encoded']

print("üìä Kolom numerik yang akan distandarisasi:")
print(numerical_cols)

scaler = StandardScaler()
df_scaled = df.copy()
df_scaled[numerical_cols] = scaler.fit_transform(df_scaled[numerical_cols])

print("\n‚úÖ Standarisasi selesai!")

X = df_scaled[numerical_cols]
y = df_scaled['diagnosis_encoded']

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    stratify=y,
    random_state=42
)

print("\nüìà Jumlah data latih:", X_train.shape[0])
print("üìâ Jumlah data uji:", X_test.shape[0])
print("\nDistribusi kelas pada data latih:")
print(y_train.value_counts(normalize=True))
print("\nDistribusi kelas pada data uji:")
print(y_test.value_counts(normalize=True))
```
