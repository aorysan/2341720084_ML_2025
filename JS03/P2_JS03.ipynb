```python
import pandas as pd
import pandas as pd
import io
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
```

```python
dpath = 'Titanic-Dataset-fixed.csv'
df0 = pd.read_csv(dpath)
df0.head()
```

```python
# --- 2. Encoding fitur 'Sex' ---
df0['Sex'] = df0['Sex'].map({'male': 1, 'female': 0})
print("=== Hasil Setelah Encoding 'Sex' ===")
print(df0[['Sex']].head(), "\n")
```

```python
# --- 3. Encoding fitur 'Cabin' ---
df0['Cabin'] = df0['Cabin'].fillna('Unknown')
print("=== Hasil Setelah Fill NaN pada 'Cabin' ===")
print(df0[['Cabin']].head(5), "\n")
df0['Cabin'] = df0['Cabin'].apply(lambda x: x[0] if x != 'Unknown' else 'U')
print("=== Hasil Ambil huruf pertama 'Cabin' (Deck) ===")
print(df0[['Cabin']].head(5), "\n")
```

```python
# --- 4. Standarisasi fitur 'Age' ---
df0['Age'] = df0[['Age']].fillna(df0['Age'].median())
print("=== Hasil Setelah Imputasi Data pada Usia dengan median ===")
print(df0[['Age']].head(10), "\n")
```

```python
scaler = StandardScaler()
df0['Age'] = scaler.fit_transform(df0[['Age']])
print("=== Hasil Setelah Standardisasi 'Age' ===")
print(df0[['Age']].head(10), "\n")
```
